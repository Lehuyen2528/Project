```python
df.drop('index', axis=1, inplace= True)
df.head()
```

 - check for duplicate rows and drop duplicate rows except 1st row

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(inplace=True)
df.shape
```

- Remove number from company name

```python
df['Company Name'] = df['Company Name'].apply(lambda x: x.split("\n")[0])
```

```python
# Convert the salary column into intergers:
df['Salary Estimate'].value_counts()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(\$|K|\s\(Glassdoor est.\)|\s\(Employer est.\))', '', regex=True)
df['Salary Estimate'] = df['Salary Estimate'].str.extract('(\d+-\d+)', expand=False)
df['Salary Estimate'] 
```

```python
# Extract Min and Max Salary
df['Min_Salary'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[0]))
df['Max_Salary'] = df['Salary Estimate'].apply(lambda x: int(x.split('-')[1]))
df['Avg_salary'] = df[['Min_Salary', 'Max_Salary']].mean(axis=1)
df[['Min_Salary', 'Max_Salary','Avg_salary']]
```

- Extracting State from Location Column and replacing Full Names with appropriate state code 

```python
df['Location'].apply(lambda x: x.split(",")[-1]).value_counts()
```

```python
df['Job_State'] = df['Location'].apply(lambda x: x.split(",")[-1].strip())
df['Job_State'].replace(['United States', 'Texas','California','New Jersey','Remote','Utah'],['US','TX','CA','NJ', np.nan,'UT'], inplace=True)
df['Job_State']
```

- Comparing job_state and Headquaerters location

```python
df['Headquarters1']= df['Headquarters'].apply(lambda x: x.split(',')[-1].strip())
df['Same_state_w_Head'] = np.where(df['Headquarters1'] == df['Job_State'],'Yes','No')
df.drop('Headquarters1', axis= 1, inplace= True)
df.head()
```

- Calculating company Age from Founded Year

```python
df['Founded'] = np.where(df['Founded'] == -1, 0, df['Founded'])
current_year = datetime.datetime.now().year
df['company_age'] = np.where(df['Founded'] == 0, np.nan, current_year - df['Founded'])
df['company_age']
```

```python
df.Rating = np.where(df.Rating == -1, 0, df.Rating)
df.sort_values(by = 'Rating', ascending= False, inplace=True)
df
```

- extracting key skills mentioned in the job description

```python
df['Job Description'][0].split('\n\n')
```

```python
df['Python'] = df['Job Description'].apply(lambda x: 'Y' if 'python' in x.lower() else 'N')
df['Excel'] = df['Job Description'].apply(lambda x: 'Y' if 'excel' in x.lower() else 'N')
df['hadoop'] = df['Job Description'].apply(lambda x: 'Y' if 'hadoop' in x.lower() else 'N')
df['spark'] = df['Job Description'].apply(lambda x: 'Y' if 'spark' in x.lower() else 'N')
df['aws'] = df['Job Description'].apply(lambda x: 'Y' if 'aws' in x.lower() else 'N')
df['tableau'] = df['Job Description'].apply(lambda x: 'Y' if 'tableau' in x.lower() else 'N')
df['big_data'] = df['Job Description'].apply(lambda x: 'Y' if 'big_data' in x.lower() else 'N')
df.head()
```
